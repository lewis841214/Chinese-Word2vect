```python
from __future__ import division
import numpy as np
import operator

def subsamling(filename):
    def count_word(filename):
        dict={}
        c=0
        number_of_word=0
        original_file=open(filename,encoding="utf8")
        for line in original_file:
            #print(line.split())
            #print(type(line.split()))
            if not line.split():
                c=c+1
            else:
                try:
                    number_of_word=number_of_word+1
                    dict[line.split()[0]]=dict[line.split()[0]]+1
                except:
                   dict[line.split()[0]]=1
                   number_of_word=number_of_word+1
        return dict,number_of_word

    def subsamplingtofile(filename,word_dict,number_of_word):
        original_file=open(filename,encoding="utf8")
        sub_sample_file=open(filename[:-4]+"_subsample"+".txt",'w',encoding="utf8")
        dict_keepingp={}
        #create dict_keepingp by ((zwi/0.001+1)**0.5*0.001/zwi)
        for key,val in  word_dict.items():
            zwi=word_dict[key]/number_of_word
            #print(zwi)
            #smaller the g, more words will be deleted
            g=0.0001
            if(((zwi/g+1)**0.5*g/zwi)>1):
                dict_keepingp[key]=1
            else:
                dict_keepingp[key]=((zwi/g+1)**0.5*g/zwi)
        c=0
        for line in original_file:
            #print(line.split())
            #print(type(line.split()))
            if not line.split():
                c=c+1
            else:
                p=dict_keepingp[line.split()[0]]
                if(np.random.uniform(0,1)<p):
                    sub_sample_file.write(line.split()[0])
                    sub_sample_file.write('\n')
        original_file.close()
        sub_sample_file.close()   
        
    def print_top_k_word_of_dict(word_dict,k):
        sorted_x = sorted(word_dict.items(), key=operator.itemgetter(1), reverse=True)
        for i in range(k):
            print(sorted_x[i])
    
    filename=filename
    word_dict,number_of_word=count_word(filename)
    print_top_k_word_of_dict(word_dict,10)
    subsamplingtofile(filename,word_dict,number_of_word)
    word_dict,number_of_word=count_word(filename[:-4]+"_subsample1"+".txt")
    print("after subsampling")
    print_top_k_word_of_dict(word_dict,10)
    
    
def count_word(filename):
    dict={}
    c=0
    number_of_word=0
    original_file=open(filename,encoding="utf8")
    for line in original_file:
        #print(line.split())
        #print(type(line.split()))
        if not line.split():
            c=c+1
        else:
            try:
                number_of_word=number_of_word+1
                dict[line.split()[0]]=dict[line.split()[0]]+1
            except:
               dict[line.split()[0]]=1
               number_of_word=number_of_word+1
    return dict,number_of_word
def print_top_k_word_of_dict(word_dict,k):
        sorted_x = sorted(word_dict.items(), key=operator.itemgetter(1), reverse=True)
        for i in range(k):
            print(sorted_x[i])
            
def merge_file():  
    merged=open("merged_wiki_text_with_only_utf8_characters_cut.txt",'w',encoding="utf8")
    file=[]
    c=7
    d=0
    for i in range(c):
        #or in ./data/....
        file.append(open("wiki_text_with_only_utf8_characters_%d_cut.txt"%(i),encoding="utf8"))
        print(file[i])
    for i in range(c):
        for line in file[i]:
            merged.write(line)
    #file.append(open("wiki_text_with_only_utf8_characters_0_cut.txt",encoding="utf8"))
    #file2=open("wiki_text_with_only_utf8_characters_1_cut.txt",encoding="utf8")
    #file3=open("wiki_text_with_only_utf8_characters_2_cut.txt",encoding="utf8")
    print(d)
    merged.close()
    for i in range(c):
        file[i].close()
```
